```python
import requests, datetime
def check_bucket(name):
    url = f"https://{name}.s3.amazonaws.com/"
    r = requests.get(url, timeout=5)
    return r.status_code

buckets = [
    "skct-727823tucy048-public",
    "skct-727823tucy048-private",
    "skct-727823tucy048-dev"
]
for b in buckets:
    code = check_bucket(b)
    print(f"{b} → {code}")
```
